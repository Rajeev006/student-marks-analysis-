```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Name': ['Raj', 'Aman', 'Sita', 'Priya', 'Vikram'],
    'Maths': [85, 78, 92, 74, 88],
    'Science': [91, 82, 89, 76, 85],
    'English': [78, 75, 94, 80, 82]
}

df = pd.DataFrame(data)

df['Average'] = df[['Maths', 'Science', 'English']].mean(axis=1)
print("Topper:")
print(df.loc[df['Average'].idxmax()])

sns.barplot(x='Name', y='Average', data=df)
plt.title("Average Marks of Students")
plt.show()
```

# 📊 Student Marks Analysis Project

This project analyzes the marks of 5 students in Maths, Science, and English.

- 📌 Calculates average
- 🏆 Finds topper
- 📈 Visualizes average marks
